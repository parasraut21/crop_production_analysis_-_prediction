# **Normalization**

```python
from sklearn.preprocessing import MinMaxScaler

columns_to_normalize = ['maxtempC', 'mintempC', 'humidity', 'precipMM', 'winddirDegree', 'windspeedKmph', 'arrivals_in_qtl', 'min_price', 'max_price', 'modal_price']

normalized_data = pune_merge_data.copy()

scaler = MinMaxScaler()
normalized_data[columns_to_normalize] = scaler.fit_transform(normalized_data[columns_to_normalize])
normalized_data
```

```python
import matplotlib.pyplot as plt
import pandas as pd

yearly_production = normalized_data.groupby(['Year', 'Commodity'])['arrivals_in_qtl'].sum().reset_index()

unique_crops = yearly_production['Commodity'].unique()
colors = plt.cm.viridis_r.colors[:len(unique_crops)]  # You can choose any colormap

colormap = dict(zip(unique_crops, colors))

plt.figure(figsize=(12, 8))

for crop, color in colormap.items():
    crop_data = yearly_production[yearly_production['Commodity'] == crop]
    plt.bar(crop_data['Year'], crop_data['arrivals_in_qtl'], label=crop, color=color)

    max_index = crop_data['arrivals_in_qtl'].idxmax()
    max_value = crop_data.loc[max_index, 'arrivals_in_qtl']
    plt.annotate(f'{crop}', xy=(crop_data.loc[max_index, 'Year'], max_value),
                 xytext=(5, 5), textcoords='offset points', color=color, fontsize=8, ha='center')

plt.xlabel('Year')
plt.ylabel('Total Crop Production (quintals)')
plt.title('Year-wise Crop Production by Crop Type')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

In 2015 Methi Bhaji, Mint were highest , and in 2016 Onion was highest

RandomForest

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer

# Assume X contains relevant weather features and y is the target variable 'Commodity'
features = ['maxtempC', 'mintempC', 'humidity', 'precipMM', 'winddirDegree', 'windspeedKmph']
X = normalized_data[features]
y = normalized_data['Commodity']

# Handle missing values
imputer = SimpleImputer(strategy='mean')
X_imputed = imputer.fit_transform(X)

# Scale numerical features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_imputed)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Choose the RandomForestClassifier as an example model
model = RandomForestClassifier(n_estimators=100, random_state=42)

# Train the model
model.fit(X_train, y_train)

# Make predictions on the testing set
y_pred = model.predict(X_test)

# Evaluate the model
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))

# Cross-validation
cv_scores = cross_val_score(model, X_scaled, y, cv=5)
print("Cross-Validation Scores:", cv_scores)
print("Mean Cross-Validation Score:", cv_scores.mean())
```

Accuracy:

The accuracy of the model is very low (approximately 1.77%). This indicates that the model is not performing well in correctly classifying the crops.

Insights :
The low accuracy, precision, recall, and F1-scores suggest that the model is struggling to generalize and predict crop types accurately based on the provided weather features
